```python
import matplotlib.pyplot as plt
import csv


x = []
y = []

with open('G:\\Andrii\\SEC\\10-17-2017 8739 SKF BEFORE CONC.CSV', 'r', encoding='utf-16') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x.append(float(row[0]))
        y.append(float(row[1]))
xx = []
yy = []
with open('G:\\Andrii\\SEC\\10-09-2017 8739 SKF 4L.CSV', 'r', encoding='utf-16') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        xx.append(float(row[0]))
        yy.append(float(row[1]))

plt.plot(xx,yy, label='Another curve')
plt.plot(x,y, label='GB1 bound to SKF')
plt.ylim(-1, 15)
plt.xlabel('Time, min')
plt.ylabel('mAU')
plt.title('aSEC')
plt.legend()
plt.show()
```
